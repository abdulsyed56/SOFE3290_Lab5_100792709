##5. Set Up Great Expectations Context and Data Source

```python
context = gx.get_context()

datasource_name = "labels_datasource"
asset_name = "labels_asset"

try:
    data_source = context.data_sources.get(datasource_name)
except Exception:
    data_source = context.data_sources.add_pandas(name=datasource_name)

try:
    data_asset = data_source.get_asset(asset_name)
except Exception:
    data_asset = data_source.add_dataframe_asset(name=asset_name)
```

##6. Define and Create a Data Batch

```python
batch_definition_name = "labels_batch_definition"

try:
    batch_definition = data_asset.get_batch_definition(batch_definition_name)
except Exception:
    batch_definition = data_asset.add_batch_definition_whole_dataframe(
        batch_definition_name
    )

batch = batch_definition.get_batch(batch_parameters={"dataframe": df})
batch
```

### Expectation 1

```python
expectation_1 = gx.expectations.ExpectColumnValuesToBeUnique(
    column="Timestamp"
)
expectation_1
```

### Validate Data Against Expectation 1

```python
result_1 = batch.validate(expectation_1)
result_1
```

### Expectation 2

```python
expectation_2 = gx.expectations.ExpectColumnValuesToNotBeNull(
    column="Ground_Truth_View"
)
expectation_2
```

### Validate Data Against Expectation 2

```python
result_2 = batch.validate(expectation_2)
result_2
```

### Expectation 3

```python
expectation_3 = gx.expectations.ExpectColumnValuesToMatchRegex(
    column="Occluded_Image_view",
    regex=r"^A_\d{3}\.png$"
)
expectation_3
```

### Validate Data Against Expectation 3

```python
result_3 = batch.validate(expectation_3)
result_3
```

Expectation 1 checks that each timestamp is unique. This is useful because each row appears to represent a separate capture/event, so duplicate timestamps could indicate duplicated records or logging problems.

Expectation 2 checks that Ground_Truth_View is never missing. This matters because the ground-truth image is a key reference in the dataset, and missing values would make comparison or validation unreliable.

Expectation 3 checks that Occluded_Image_view follows the expected filename format A_###.png. This helps confirm that the image naming is consistent and that the file references were recorded properly.
